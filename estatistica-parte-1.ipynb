```python
dados = pd.DataFrame(data = siga, columns = ['DatGeracaoConjuntoDados', 'SigUFPrincipal', 'SigTipoGeracao', 'DscFaseUsina', 'DscOrigemCombustivel', 'DscFonteCombustivel', 'DatEntradaOperacao', 'MdaPotenciaOutorgadaKw', 'MdaPotenciaFiscalizadaKw', 'IdcGeracaoQualificada', 'DatInicioVigencia', 'DatFimVigencia'])
dados
```

```python
dados.MdaPotenciaOutorgadaKw = dados.MdaPotenciaOutorgadaKw.str.replace(",", ".").astype(float)
dados.DatGeracaoConjuntoDados = pd.to_datetime(dados.DatGeracaoConjuntoDados)
dados.DatEntradaOperacao = pd.to_datetime(dados.DatEntradaOperacao)
dados.DatInicioVigencia = pd.to_datetime(dados.DatInicioVigencia)
dados.DatFimVigencia = pd.to_datetime(dados.DatFimVigencia)
dados.info()
```

```python
from datetime import date

dados['Tempo_de_operacao_em_dias'] = dados.DatGeracaoConjuntoDados - dados.DatEntradaOperacao
dados
```

```python
dados['Tempo_de_operacao_em_anos'] = dados['Tempo_de_operacao_em_dias']/ np.timedelta64(1, 'Y')
dados
```

```python
dados['Tempo_de_vigencia_em_dias'] = dados.DatFimVigencia - dados.DatInicioVigencia
dados['Tempo_de_vigencia_em_anos'] = dados['Tempo_de_vigencia_em_dias']/ np.timedelta64(1, 'Y')
dados
```

```python
dados['Tempo_de_operacao_em_anos'] = dados['Tempo_de_operacao_em_anos'].fillna(0.0).astype(int)
dados['Tempo_de_vigencia_em_anos'] = dados['Tempo_de_vigencia_em_anos'].fillna(0.0).astype(int)
dados
```

```python
eolica_operando = eolica.query('DscFaseUsina == "Operação"')
eolica_operando
```

```python
media_potencia = eolica_operando.MdaPotenciaOutorgadaKw.mean()
media_potencia
```

```python
eolica_operando['Desvio'] = eolica_operando['MdaPotenciaOutorgadaKw'] - media_potencia
eolica_operando
```

```python
eolica_operando['Desvio'].sum()
```

```python
eolica_operando['|Desvio|'] = eolica_operando['Desvio'].abs()
eolica_operando
```

```python
eolica_operando['|Desvio|'].mean()
```

```python
desvio_medio_absoluto = eolica_operando.MdaPotenciaOutorgadaKw.mad()
desvio_medio_absoluto
```

```python
eolica_operando['(Desvio)^2'] = eolica_operando['Desvio'].pow(2)
eolica_operando
```

```python
variancia = eolica_operando.MdaPotenciaOutorgadaKw.var()
variancia
```

```python
desvio_padrao = np.sqrt(variancia)
desvio_padrao
```
